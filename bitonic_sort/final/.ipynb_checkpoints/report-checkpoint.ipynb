```python
tk = th.Thicket.from_caliperreader(glob("./*.cali"))
```

```python
tk.metadata_column_to_perfdata("num_procs")
tk.metadata_column_to_perfdata("input_size")
tk.metadata_column_to_perfdata("input_type")

df = tk.dataframe.reset_index()
df.head()
```

```python
import math
```

```python
def plot_large_data(df, name, ax, input_type):
    df_filtered = df[(df["input_type"] == input_type) & (df["name"] == name)]
    
    df_22 = df_filtered[df_filtered["input_size"] == math.pow(2, 24)].sort_values("num_procs")
    df_24 = df_filtered[df_filtered["input_size"] == math.pow(2, 26)].sort_values("num_procs")
    df_26 = df_filtered[df_filtered["input_size"] == math.pow(2, 28)].sort_values("num_procs")
    
    df_22.plot(x="num_procs", y="Avg time/rank", ax=ax, label="Input size: 2^24")
    df_24.plot(x="num_procs", y="Avg time/rank", ax=ax, label="Input size: 2^26")
    df_26.plot(x="num_procs", y="Avg time/rank", ax=ax, label="Input size: 2^28")
    
    ax.set_xscale('log', base=2)
    ax.set_title(f"{name}")
    ax.set_xlabel("Number of Processes")
    ax.set_ylabel("Avg time/rank")
    ax.legend()
```

```python
fig, (ax1, ax2, ax3) = plt.subplots(1, 3, figsize=(20, 6))

plot_large_data(df, "comm", ax1, "random")
plot_large_data(df, "comp_large", ax2, "random")
plot_large_data(df, "main", ax3, "random")

plt.tight_layout()
plt.show()
```

```python
fig, (ax1, ax2, ax3) = plt.subplots(1, 3, figsize=(20, 6))

plot_large_data(df, "comm", ax1, "sorted")
plot_large_data(df, "comp_large", ax2, "sorted")
plot_large_data(df, "main", ax3, "sorted")

plt.tight_layout()
plt.show()
```

```python
fig, (ax1, ax2, ax3) = plt.subplots(1, 3, figsize=(20, 6))

plot_large_data(df, "comm", ax1, "reverse_sorted")
plot_large_data(df, "comp_large", ax2, "reverse_sorted")
plot_large_data(df, "main", ax3, "reverse_sorted")

plt.tight_layout()
plt.show()
```

```python
fig, (ax1, ax2, ax3) = plt.subplots(1, 3, figsize=(20, 6))

plot_large_data(df, "comm", ax1, "one_percent_pertubed")
plot_large_data(df, "comp_large", ax2, "one_percent_pertubed")
plot_large_data(df, "main", ax3, "one_percent_pertubed")

plt.tight_layout()
plt.show()
```

```python
def plot_small_data(df, name, ax, input_type):
    df_filtered = df[(df["input_type"] == input_type) & (df["name"] == name)]
    
    df_16 = df_filtered[df_filtered["input_size"] == math.pow(2, 16)].sort_values("num_procs")
    df_18 = df_filtered[df_filtered["input_size"] == math.pow(2, 18)].sort_values("num_procs")
    df_20 = df_filtered[df_filtered["input_size"] == math.pow(2, 20)].sort_values("num_procs")
    
    df_16.plot(x="num_procs", y="Avg time/rank", ax=ax, label="Input size: 2^16")
    df_18.plot(x="num_procs", y="Avg time/rank", ax=ax, label="Input size: 2^18")
    df_20.plot(x="num_procs", y="Avg time/rank", ax=ax, label="Input size: 2^20")
    
    ax.set_xscale('log', base=2)
    ax.set_title(f"{name}")
    ax.set_xlabel("Number of Processes")
    ax.set_ylabel("Avg time/rank")
    ax.legend()
```

```python
fig, (ax1, ax2, ax3) = plt.subplots(1, 3, figsize=(20, 6))

plot_small_data(df, "comm", ax1, "random")
plot_small_data(df, "comp_large", ax2, "random")
plot_small_data(df, "main", ax3, "random")

plt.tight_layout()
plt.show()
```

```python
fig, (ax1, ax2, ax3) = plt.subplots(1, 3, figsize=(20, 6))

plot_small_data(df, "comm", ax1, "sorted")
plot_small_data(df, "comp_large", ax2, "sorted")
plot_small_data(df, "main", ax3, "sorted")

plt.tight_layout()
plt.show()
```
