```python
import math
```

```python
def plot_large_data(df, name, ax, input_type):
    df_filtered = df[(df["input_type"] == input_type) & (df["name"] == name)]
    
    df_22 = df_filtered[df_filtered["input_size"] == math.pow(2, 24)].sort_values("num_procs")
    df_24 = df_filtered[df_filtered["input_size"] == math.pow(2, 26)].sort_values("num_procs")
    df_26 = df_filtered[df_filtered["input_size"] == math.pow(2, 28)].sort_values("num_procs")
    
    df_22.plot(x="num_procs", y="Avg time/rank", ax=ax, label="Input size: 2^24")
    df_24.plot(x="num_procs", y="Avg time/rank", ax=ax, label="Input size: 2^26")
    df_26.plot(x="num_procs", y="Avg time/rank", ax=ax, label="Input size: 2^28")
    
    ax.set_xscale('log', base=2)
    ax.set_title(f"{name}")
    ax.set_xlabel("Number of Processes")
    ax.set_ylabel("Avg time/rank")
    ax.legend()
```

```python
fig, (ax1, ax2, ax3) = plt.subplots(1, 3, figsize=(20, 6))

plot_large_data(df, "comm", ax1, "random")
plot_large_data(df, "main", ax3, "random")
plot_large_data(df, "comp_large", ax2, "random")

plt.tight_layout()
plt.show()
```

```python
fig, (ax1, ax2, ax3) = plt.subplots(1, 3, figsize=(20, 6))

plot_large_data(df, "comm", ax1, "sorted")
plot_large_data(df, "comp_large", ax2, "sorted")
plot_large_data(df, "main", ax3, "sorted")

plt.tight_layout()
plt.show()
```

```python
fig, (ax1, ax2, ax3) = plt.subplots(1, 3, figsize=(20, 6))

plot_large_data(df, "comm", ax1, "reverse_sorted")
plot_large_data(df, "comp_large", ax2, "reverse_sorted")
plot_large_data(df, "main", ax3, "reverse_sorted")

plt.tight_layout()
plt.show()
```

```python
fig, (ax1, ax2, ax3) = plt.subplots(1, 3, figsize=(20, 6))

plot_large_data(df, "comm", ax1, "perturbed")
plot_large_data(df, "comp_large", ax2, "perturbed")
plot_large_data(df, "main", ax3, "perturbed")

plt.tight_layout()
plt.show()
```

```python
def algorithm_comparison(df, ax, array_size, name):
    df_filtered = df[(df["input_size"] == math.pow(2, array_size)) & (df["name"] == name)]

    implementations = ["random", "sorted", "reverse_sorted", "one_percent_pertubed"]

    for i in implementations:
        df_ = df_filtered[df_filtered["input_type"] == i].sort_values("num_procs")
        df_.plot(x="num_procs", y="Avg time/rank", ax=ax, label=f"{i}")
    
    ax.set_xscale('log', base=2)
    ax.set_title(f"{name} time for Array size: 2^{array_size}")
    ax.set_xlabel("Number of Processes")
    ax.set_ylabel("Avg time/rank")
    ax.legend()
```

```python
# Comparing runtimes of different input types

fig, (ax1, ax2, ax3) = plt.subplots(1, 3, figsize=(20, 6))

algorithm_comparison(df, ax1, 28, "comm")
algorithm_comparison(df, ax2, 28, "comp_large")
algorithm_comparison(df, ax3, 28, "main")

plt.tight_layout()
plt.show()
```

```python
fig, (ax1, ax2, ax3) = plt.subplots(1, 3, figsize=(20, 6))

algorithm_comparison(df, ax1, 16, "comm")
algorithm_comparison(df, ax2, 16, "comp_large")
algorithm_comparison(df, ax3, 16, "main")

plt.tight_layout()
plt.show()
```
