```python
def plot_large_data(df, name, ax, input_type):
    df_filtered = df[(df["input_type"] == input_type) & (df["name"] == name)]
    
    df_22 = df_filtered[df_filtered["input_size"] == math.pow(2, 24)].sort_values("num_procs")
    df_24 = df_filtered[df_filtered["input_size"] == math.pow(2, 26)].sort_values("num_procs")
    df_26 = df_filtered[df_filtered["input_size"] == math.pow(2, 28)].sort_values("num_procs")
    
    df_22.plot(x="num_procs", y="Max time/rank", ax=ax, label="Input size: 2^24")
    df_24.plot(x="num_procs", y="Max time/rank", ax=ax, label="Input size: 2^26")
    df_26.plot(x="num_procs", y="Max time/rank", ax=ax, label="Input size: 2^28")
    
    ax.set_xscale('log', base=2)
    ax.set_title(f"{name}")
    ax.set_xlabel("Number of Processes")
    ax.set_ylabel("Max time/rank")
    ax.legend()
```

```python
fig, (ax1, ax2, ax3) = plt.subplots(1, 3, figsize=(20, 6))

plot_large_data(df, "comm", ax1, "random")
plot_large_data(df, "main", ax3, "random")
plot_large_data(df, "comp_large", ax2, "random")

plt.tight_layout()
plt.show()
```

```python
def algorithm_comparison(df, ax, array_size, name):
    df_filtered = df[(df["input_size"] == math.pow(2, array_size)) & (df["name"] == name)]

    implementations = ["random", "sorted", "reverse_sorted", "perturbed"]

    for i in implementations:
        df_ = df_filtered[df_filtered["input_type"] == i].sort_values("num_procs")
        df_.plot(x="num_procs", y="Max time/rank", ax=ax, label=f"{i}")
    
    ax.set_xscale('log', base=2)
    ax.set_title(f"{name} time for Array size: 2^{array_size}")
    ax.set_xlabel("Number of Processes")
    ax.set_ylabel("Max time/rank")
    ax.legend()
```

```python
# Comparing runtimes of different input types

fig, (ax1, ax2, ax3) = plt.subplots(1, 3, figsize=(20, 6))

algorithm_comparison(df, ax1, 28, "comm")
algorithm_comparison(df, ax2, 28, "comp_large")
algorithm_comparison(df, ax3, 28, "main")

plt.tight_layout()
plt.show()
```

```python
fig, (ax1, ax2, ax3) = plt.subplots(1, 3, figsize=(20, 6))

algorithm_comparison(df, ax1, 16, "comm")
algorithm_comparison(df, ax2, 16, "comp_large")
algorithm_comparison(df, ax3, 16, "main")

plt.tight_layout()
plt.show()
```

```python
# Speedup
def plot_speedup(df, ax, input_type):
    df_filtered = df[(df["input_type"] == input_type) & (df["name"] == "main")]

    array_sizes = [math.pow(2, 16), math.pow(2, 18), math.pow(2, 20), math.pow(2, 22), math.pow(2, 24), math.pow(2, 26), math.pow(2, 28)]
    for size in array_sizes:
        t1 = df_filtered[(df_filtered["num_procs"] == 2) & (df_filtered["input_size"] == size)]["Max time/rank"].max() * 2
        df_ = df_filtered[df_filtered["input_size"] == size].sort_values("num_procs")
        df_["speedup"] = t1 / df_["Max time/rank"]
        df_.plot(x="num_procs", y="speedup", ax=ax, label=f"{size}")
    
    ax.set_xscale('log', base=2)
    ax.set_title(f"Strong scaling speedup time for {input_type}")
    ax.set_xlabel("Number of Processes")
    ax.set_ylabel("Speedup")
    ax.legend()
```

```python
fig, (ax1, ax2, ax3, ax4) = plt.subplots(1, 4, figsize=(25, 6))

plot_speedup(df, ax1, "random")
plot_speedup(df, ax2, "sorted")
plot_speedup(df, ax3, "reverse_sorted")
plot_speedup(df, ax4, "perturbed")

plt.tight_layout()
plt.show()
```

```python
# weak scaling
def plot_weak_scaling(df, ax, name):
    input_types = ["random", "sorted", "reverse_sorted", "perturbed"]
    
    array_sizes = [math.pow(2, 16), math.pow(2, 18), math.pow(2, 20), 
                math.pow(2, 22), math.pow(2, 24)]
    proc_sizes = [2, 8, 32, 128, 512]

    ax.clear()
    
    for input_type in input_types:
        df_filtered = df[(df["input_type"] == input_type) & (df["name"] == name)]
        
        times = []
        procs = []
        
        for i in range(len(array_sizes)):
            df_ = df_filtered[(df_filtered["input_size"] == array_sizes[i]) & 
                            (df_filtered["num_procs"] == proc_sizes[i])]
            if not df_.empty:
                times.append(df_["Max time/rank"].values[0])
                procs.append(proc_sizes[i])
        
        ax.plot(procs, times, 'o-', label=f"{input_type}")
    
    ax.set_xscale('log', base=2)
    ax.set_title(f"Weak scaling for {name}")
    ax.set_xlabel("Number of Processes")
    ax.set_ylabel("Max time/rank")
    ax.legend()
```

```python
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(12, 6))

plot_weak_scaling(df, ax1, "main")
plot_weak_scaling(df, ax1, "main")
plot_weak_scaling(df, ax1, "main")
plot_weak_scaling(df, ax1, "main")

plot_weak_scaling(df, ax2, "comm")
plot_weak_scaling(df, ax2, "comm")
plot_weak_scaling(df, ax2, "comm")
plot_weak_scaling(df, ax2, "comm")

plt.tight_layout()
plt.show()
```
